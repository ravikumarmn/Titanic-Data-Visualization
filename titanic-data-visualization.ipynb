# Importing Required Libraries. 
In This Section We Use Plotly Library For Visualization Purpose.

```python
import pandas as pd
import plotly.express as px
from plotly.offline import plot,iplot,init_notebook_mode
init_notebook_mode(connected=True)
```

**Reading The Data.**

```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
test_df = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
train_df.head()
```

# **Let's Try to Find Solution To Few Questions.**

```python
name = train_df['Survived'].replace(0,'No').replace(1,'Yes')
px.pie(train_df,values = 'PassengerId',names = name,color = 'Survived',hover_data = ['Sex','Age'],title = 'Percentage of Passengers Survived : Survived = 1, Not Survived = 0',labels = {'x':'Survived','y':'Not Survived'})
```

**Looking at the Pie Chart we get that, 38.2% are Survied, Rest don't.**

```python
px.bar(train_df,x = 'PassengerId',y = 'Sex',color = 'Sex',title = 'Which Sex has a more strenght?')
```

```python
px.scatter(train_df,x = 'PassengerId',y = 'Fare',color = 'Sex',size='Fare',hover_data = ['Age','Sex'])
```

```python
px.histogram(train_df,x = 'Sex',y = 'Survived',color = 'Sex',title = 'Female Passengers Has Survived More Than Male').show()
px.pie(train_df,names = 'Sex',values = 'Survived',color = 'Sex',title = 'Female Passengers Has Survived More Than Male').show()
```

```python
pd.cut(train_df['Age'],bins = 4).value_counts()
```

```python
px.histogram(train_df,x = 'Age',nbins = 5,marginal = 'box',color = 'Sex')
```

```python
px.box(train_df,x = 'Pclass',y = 'Age',points = 'all',notched = True)
```

```python
px.density_heatmap(train_df,x = 'Pclass',y = 'Age',color_continuous_scale='Viridis')
```

```python
px.scatter_3d(train_df,x = 'Pclass',y = 'Age',z = 'Fare',color = 'Survived',opacity=0.7)
```
